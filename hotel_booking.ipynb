# Data Analysis And Visualization

```python
# First we will focusing to find out the percentage of the cancellation of the booking
cancelled_percentage = df['is_canceled'].value_counts(normalize=True)
cancelled_percentage
```

Here we can see that 37% customer cancelled the booking and 62% did not cancelled the booking

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not cancelled','Cancelled'],df['is_canceled'].value_counts(),color=['green','red'])
plt.xlabel('Reservations')
plt.ylabel('counts')
plt.gca().set_facecolor('darkgray')
plt.gcf().set_facecolor('darkgray')
plt.show()
```

```python
plt.figure(figsize=(4, 4), facecolor='darkgray')  # Change the background color of the figure
ax1 = plt.gca()  # Get current axes
ax1.set_facecolor('darkgray')  # Change the background color of the axes

# Use seaborn to create the countplot
sns.countplot(x='hotel', hue='is_canceled', data=df, palette=['green', 'red'])

plt.title('Reservation status in different hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')

# Adjust the legend
legend_labels = ['Not canceled', 'Canceled']
plt.legend(title='Status', labels=legend_labels, bbox_to_anchor=(1.05, 1))

plt.show()
```

Now take a look at both resort and city hotel percentage of cenled and not canceld data individualliy

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(25,10))
plt.plot(resort_hotel.index,resort_hotel['adr'], label='resort_hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label='city_hotel')
plt.title('Average daily rate in city and resort hotels', fontsize=20)
plt.legend(fontsize= 20)
plt.gca().set_facecolor('black')
plt.gcf().set_facecolor('darkgray')
plt.tick_params(axis='x', color='white')
  
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled','canceled'])
plt.gca().set_facecolor('darkgray')
plt.gcf().set_facecolor('darkgray')
plt.show()
```

```python
# Group by 'reservation_status_date' and calculate the mean of 'adr' for cancelled data
cancelled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Filter for not cancelled data
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the data
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate')
plt.legend()
plt.show()
```

```python
# Corrected filter for cancelled_df_adr
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016-01-01') & (cancelled_df_adr['reservation_status_date'] < '2017-09-30')]

# Corrected filter for not_cancelled_df_adr
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016-01-01') & (not_cancelled_df_adr['reservation_status_date'] < '2017-09-30')]

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate')
plt.legend()
plt.show()
```
